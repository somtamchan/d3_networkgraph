```python
import pandas as pd
import itertools
```

```python
df = pd.read_csv("JPpatent_kidney_disease_2019.csv")
```

```python
fi_list = [l.split(",")for l in list(df["FI"])]
```

```python
result_list= []
for l in fi_list:
    tmp = []
    for t in l:
        fi_split = t.split("/")
        if len(fi_split) == 2:
            tmp.append(fi_split[0])
        result = list(set(tmp))
    result_list.append(result)
```

```python
doc_id_list = []
for i in range(df.shape[0]):
    doc_id = {}
    doc_id["id"] = i
    doc_id["label"] = "patent" + str(i)
    doc_id_list.append(doc_id)
```

```python
fi_dic = {}
fi_id_list = []
fi_list_flatten = list(set(list(itertools.chain.from_iterable(result_list))))
for j, fi in enumerate(fi_list_flatten):
    fi_id = {}
    fi_dic[fi] = df.shape[0] + j
    fi_id["id"] = df.shape[0] + j
    fi_id["label"] = fi 
    fi_id_list.append(fi_id)
```

```python
link_list = []
for i, r in enumerate(result_list):
    for fi in r:
        link_dic = {}
        link_dic["source"] = i
        link_dic["target"] = fi_dic[fi]
        link_list.append(link_dic)
        print(link_dic)
```
